```python
import random
import pandas as pd
from faker import Faker
import numpy as np

# Initialize Faker and random seed for reproducibility
fake = Faker()
Faker.seed(42)
random.seed(42)

# Define list of locations and states in India
locations = ['Mumbai', 'Delhi', 'Kolkata', 'Bangalore', 'Chennai', 'Hyderabad', 'Pune', 'Jaipur', 'Lucknow', 'Ahmedabad']
states = ['Maharashtra', 'Delhi', 'West Bengal', 'Karnataka', 'Tamil Nadu', 'Telangana', 'Uttar Pradesh', 'Gujarat', 'Rajasthan', 'Punjab']

# Define legitimate UPI IDs with their phone numbers
legitimate_users = {
    "Pranjal Bhinge": ("pranjalbhinge@oksbi", "8431212363"),
    "Shruti Basarikatti": ("shrutib9393@okhdfc", "8197739843"),
    "Ranjeeta Chikkaparappa": ("ranjeeta.rcb@oksbi", "8885559999"),
    "Rohan Pawar": ("8748030185@upi", "8748030185")
}

# Define function to generate a single transaction
def generate_transaction(is_fraudulent):
    # Select random legitimate user or generate a fake user
    if is_fraudulent:
        name = fake.name()
        upi_id = fake.unique.email()
        phone_number = fake.phone_number()
    else:
        name, upi_id, phone_number = random.choice(list(legitimate_users.values()))

    # Generate a random transaction amount between 1 and 10 lakh (safe range for legitimate)
    if not is_fraudulent:
        amount = random.randint(1, 1000000)
    else:
        # Fraudulent transactions will have amounts outside the expected range or randomly chosen
        amount = random.randint(1, 1000000)

    # Randomize transaction time within a reasonable time span
    transaction_time = fake.date_this_year()

    # Select random location and state
    location = random.choice(locations)
    state = random.choice(states)

    # Receiver's UPI ID (can be fake or real)
    receiver_upi_id = fake.email()

    # Return the transaction details as a dictionary
    return {
        "Sender's Identity": name,
        "Sender's UPI ID": upi_id,
        "Sender's Phone Number": phone_number,
        "Transaction Amount": amount,
        "Receiver's UPI ID": receiver_upi_id,
        "Time of Transaction": transaction_time,
        "Location": location,
        "State": state,
        "Fraudulent": 1 if is_fraudulent else 0
    }

# Create a list of transactions with a 50-50 fraudulent/non-fraudulent split
transactions = []
for _ in range(1000):  # Create 1000 fraudulent and 1000 safe transactions
    transactions.append(generate_transaction(is_fraudulent=True))  # Fraudulent
    transactions.append(generate_transaction(is_fraudulent=False))  # Safe

# Convert the list of transactions into a DataFrame
df = pd.DataFrame(transactions)

# Save the DataFrame to a CSV file
df.to_csv('upi_transaction_data.csv', index=False)

# Output sample data to verify
print(df.head())
```

```python
# Install Faker if not installed
!pip install faker

# Import libraries
import random
import pandas as pd
from faker import Faker
import numpy as np

# Initialize Faker
fake = Faker()
Faker.seed(42)
random.seed(42)

# Define Indian cities and states
locations = ['Mumbai', 'Delhi', 'Kolkata', 'Bangalore', 'Chennai', 'Hyderabad', 'Pune', 'Jaipur', 'Lucknow', 'Ahmedabad']
states = ['Maharashtra', 'Delhi', 'West Bengal', 'Karnataka', 'Tamil Nadu', 'Telangana', 'Uttar Pradesh', 'Gujarat', 'Rajasthan', 'Punjab']

# Your given fixed legitimate users
fixed_legitimate_users = {
    "Pranjal Bhinge": ("pranjalbhinge@oksbi", "8431212363"),
    "Shruti Basarikatti": ("shrutib9393@okhdfc", "8197739843"),
    "Ranjeeta Chikkaparappa": ("ranjeeta.rcb@oksbi", "8885559999"),
    "Rohan Pawar": ("8748030185@upi", "8748030185")
}

# Generate additional random legitimate users
def generate_random_legitimate_users(n):
    random_legit_users = {}
    for _ in range(n):
        name = fake.unique.name()
        upi_id = fake.user_name() + "@ok" + random.choice(["sbi", "hdfc", "icici", "axis", "paytm"])
        phone_number = fake.random_number(digits=10, fix_len=True)
        random_legit_users[name] = (upi_id, str(phone_number))
    return random_legit_users

# Merge fixed users and random users
num_random_legit_users = 500  # You can adjust how many random legitimate users you want
all_legitimate_users = {**fixed_legitimate_users, **generate_random_legitimate_users(num_random_legit_users)}

# Function to generate a transaction
def generate_transaction(is_fraudulent):
    if is_fraudulent:
        name = fake.name()
        upi_id = fake.email()
        phone_number = fake.phone_number()
    else:
        name, (upi_id, phone_number) = random.choice(list(all_legitimate_users.items()))

    amount = random.randint(1, 1000000)
    transaction_time = fake.date_this_year()
    location = random.choice(locations)
    state = random.choice(states)
    receiver_upi_id = fake.email()

    return {
        "Sender's Identity": name,
        "Sender's UPI ID": upi_id,
        "Sender's Phone Number": phone_number,
        "Transaction Amount": amount,
        "Receiver's UPI ID": receiver_upi_id,
        "Time of Transaction": transaction_time,
        "Location": location,
        "State": state,
        "Fraudulent": 1 if is_fraudulent else 0
    }

# Create the full dataset
transactions = []
for _ in range(2000):  # 1000 fraudulent + 1000 non-fraudulent = 2000 total
    transactions.append(generate_transaction(is_fraudulent=True))
    transactions.append(generate_transaction(is_fraudulent=False))

# Convert to DataFrame
df = pd.DataFrame(transactions)

# Save as CSV
df.to_csv('upi_transaction_data.csv', index=False)

# Show sample
df.head()
```
